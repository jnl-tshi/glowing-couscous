**Concept:** The core idea is to understand how a function's output changes in response to changes in its input(s). This 'change' is captured by derivatives.

Derivative:
- tells us the "instantaneous" rate of change of f(x) with respect to x.
- geometrically, it's the slope of the tangent line to the function at a given point.

**Basic derivative rules**

| Function $f(x)$           | Derivative $\frac{df}{dx}$               | Notes                                      |
| :------------------------ | :--------------------------------------- | :----------------------------------------- |
| $c$                       | $0$                                      | Constant Rule                              |
| $x^n$                     | $nx^{n-1}$                               | Power Rule                                 |
| $e^x$                     | $e^x$                                    | Exponential Rule                           |
| $\ln(x)$                  | $\frac{1}{x}$                            | Natural Logarithm Rule                     |
| $\sin(x)$                 | $\cos(x)$                                | Sine Rule                                  |
| $\cos(x)$                 | $-\sin(x)$                               | Cosine Rule                                |
| $c \cdot g(x)$            | $c \cdot g'(x)$                          | Constant Multiple Rule                     |
| $g(x) \pm h(x)$           | $g'(x) \pm h'(x)$                        | Sum/Difference Rule                        |
| $g(x) \cdot h(x)$         | $g'(x)h(x) + g(x)h'(x)$                  | Product Rule                               |
| $\frac{g(x)}{h(x)}$       | $\frac{g'(x)h(x) - g(x)h'(x)}{[h(x)]^2}$ | Quotient Rule (for completeness)           |

## Partial derivatives and gradients

Partial derivative: https://www.mathsisfun.com/calculus/derivatives-partial.html

**Concept:** When dealing with functions that have multiple input variables (like a loss function that depends on many model parameters), we use partial derivatives to understand how the function changes with respect to each variable individually.

- **A partial derivative** tells us the instantaneous rate of change of a multivariable function f as only variable x_i changes, while all other variables are held constant.

- **The gradient** of a multivariable function is a vector containing all its partial derivatives.

For a function $f(x_1,x_2,â€¦,x_n)$, its gradient is:

$$\nabla f(x_1, x_2, \dots, x_n) = \begin{pmatrix} \frac{\partial f}{\partial x_1} \\ \frac{\partial f}{\partial x_2} \\ \vdots \\ \frac{\partial f}{\partial x_n} \end{pmatrix}$$

The gradient vector points in the direction of the steepest increase (uphill) of the function at a given point. If we want to move towards a minimum (downhill), we should move in the opposite direction of the gradient.

**Machine Learning relevance**:

In Machine Learning, our goal is often to minimise a loss function, which quantifies the error of our model. This loss function typically depends on many parameters (e.g., weights and biases). The gradient of this loss function indicates how to adjust each parameter to minimise the error most efficiently. Moving against the gradient is the core principle of Gradient Descent, which we'll cover soon.

**Example:**

Let's find the gradient for the function $$f(x, y) = x^2 + 2y^2$$

1. Partial derivative with respect to x (treat y as a constant): $$\frac{\partial f}{\partial x} = 2x$$

2. Partial derivative with respect to y (treat x as a constant): $$\frac{\partial f}{\partial y} = 4y$$

3. Form the Gradient Vector: $$\nabla f(x, y) = \begin{pmatrix} 2x \\ 4y \end{pmatrix}$$

## The chain rule

**Concept:** The Chain Rule is a rule in calculus for differentiating composite functions. A composite function is a "function of a function" â€“ where the output of one function becomes the input to another.

**If** $y$ is a function of $u$

**$y=f(u)$**

**and** $u$ is itself a function of $x$

**$u=g(x)$**

**then** $y$ is indirectly a function of $x$

**$y=f(g(x))$**

The Chain Rule states how to find the derivative of $y$ with respect to $x$:

$$\frac{dy}{dx} = \frac{dy}{du} â‹… \frac{du}{dx}$$

This rule can be extended for functions composed of many nested functions. For instance, if **$y=f(g(h(x)))$**, then **$\frac{dy}{dx} = \frac{dy}{df} â‹… \frac{df}{dy} â‹… \frac{dg}{dh}$**.

**Machine Learning relevance**: Backpropagation in Neural Networks

**Example:**

Let $y=u^3$ and $u=2x+1$. We want to find  $\frac{dx}{dy}$.

1. Find $\frac{dy}{du}$:
$$\frac{dy}{du} = \frac{d}{du}(u^3) = 3u^2$$

2. Find $\frac{du}{dx}$:
$$\frac{du}{dx} = \frac{d}{dx}(2x+1) = 2$$

3. Apply the Chain Rule:
$$\frac{dy}{dx} = \frac{dy}{du} â‹… \frac{du}{dx} = 3u^2 â‹… 2 = 6u^2$$

4. Substitute $u$ back in terms of $x$:
$$\frac{dy}{dx} = 6(2x+1)^2$$

## Gradient Descent

Gradient Descent is an algorithm that can find the minimum of a function by optimising it through multiple iterations.

Gradient Descent will use partial derivative of the loss function with respect to m.

Gradient Descent use one more variable which is called **Learning Rate**. It controls how much the value of **m** changes with each step. As lower is the Learning rate as better accuracy we can get, but this means more iterations, for a higher value of learning rate, there is a risk to miss the minimum of the loss function.

Using the computed D_m, D_c and the Learning Rate (L) we update the new constants of our linear function, m and c using the following equation.

### Task: 2.1 Build a Gradient Descent function from scratch

**Your goal**: Create a Python function that performs Gradient Descent for our linear regression model: **Y=mX+c**.

**Function requirements**:

- **Function name**: Choose a descriptive name, something like *train_linear_regression_gd* or *gradient_descent_fit*.

- **Input parameters**: Your function should accept the following arguments:

3. **learning_rate**: A floating-point number that controls the step size for each update (e.g., 0.1).
4. **num_iterations**: An integer specifying how many times the m and c values should be updated.
5. **initial_m**: The starting value for the slope m (e.g., 1).
6. **initial_c**: The starting value for the intercept c (e.g., 1).

**Return Values**: Your function should return:

- The final m value.
- The final c value.
- The list of MSE values from each iteration.
- The list of m values from each iteration.
- The list of c values from each iteration.

```python
# Gradient Descent parameters
n = diamonds_subset.shape[0]
L = 0.1  # Learning rate
m = 1.0
c = 1.0
num_iterations = 300

# Lists to store history
m_values = []
c_values = []
mse_values = []

X = diamonds_subset['carat']
y = diamonds_subset['price']

#create function definition
def GDFunction(n,L,m,c,X,y):

  # Gradient Descent loop
  for _ in range(num_iterations):
      Y_pred = m * X + c
      error = y - Y_pred
      mse = np.mean(error ** 2)

      D_m = (-2/n) * sum(X * error)
      D_c = (-2/n) * sum(error)

      m = m - L * D_m
      c = c - L * D_c

      m_values.append(m)
      c_values.append(c)
      mse_values.append(mse)



  # Plot: Data and best fit lines across iterations
  plt.figure(figsize=(10, 6))
  plt.scatter(X, y, alpha=0.4, label='Data', s=10)

  x_vals = np.linspace(X.min(), X.max(), 100)
  # Define a list of colors
  colors = plt.cm.viridis(np.linspace(0, 1, num_iterations))

  for i in range(num_iterations):
      if   i % 20 == 0 or i == num_iterations - 1:
        y_vals = m_values[i] * x_vals + c_values[i]
        plt.plot(x_vals, y_vals, color=colors[i], alpha=0.3, label=f'Iter {i+1}')

  # Plot 1: m, c, and MSE over iterations
  plt.figure(figsize=(14, 5))

  plt.subplot(1, 3, 1)
  plt.plot(range(num_iterations), m_values)
  plt.title('Slope (m) over Iterations')
  plt.xlabel('Iteration')
  plt.ylabel('m value')

  plt.subplot(1, 3, 2)
  plt.plot(range(num_iterations), c_values)
  plt.title('Intercept (c) over Iterations')
  plt.xlabel('Iteration')
  plt.ylabel('c value')

  plt.subplot(1, 3, 3)
  plt.plot(range(num_iterations), mse_values)
  plt.title('MSE over Iterations')
  plt.xlabel('Iteration')
  plt.ylabel('Mean Squared Error')




  plt.tight_layout()

  return plt.show()
```

```python

# Initialize parameters for y = ln(x * a + b) * c + d
X = diamonds_subset['carat']
y = diamonds_subset['price']

n = len(X)
L1 = 0.001
L2 = 0.001
L3 = 0.001
L4 = 0.001
a, b, c, d = 0.005, 0.2, 10000.0,8000.0
#a, b, c, d = 0.05,1.0, 1.0,1.0
num_iterations = 200



# History tracking
a_values, b_values, c_values, d_values, mse_values = [], [], [], [], []

# Gradient Descent loop
for _ in range(num_iterations):
    inner = (X*a) - b
    # Avoid log domain error by clipping inner to minimum positive
    #inner = np.clip(inner, 1e-5, None)

    log_part = np.log(inner)
    y_pred = log_part * c + d
    error =  y - y_pred
    mse = np.mean(error ** 2)

    # Gradients
    D_a = (-2/n) * np.sum(X*c*(y-c*log_part -d)/(a*X+b))
    D_b = (-2/n) * np.sum(c*(y-c*(log_part)-d)/(a*X+b))
    D_c = (-2/n) * np.sum(log_part*(y-y_pred))
    D_d = (2/n) * np.sum(error)

    # Update parameters
    a -= L1 * D_a
    b -= L2 * D_b
    c -= L3 * D_c
    d -= L4 * D_d

    # Store history
    a_values.append(a)
    b_values.append(b)
    c_values.append(c)
    d_values.append(d)
    mse_values.append(mse)

# Final prediction using the learned parameters
x_vals = np.linspace(X.min(), X.max(), 100)
inner_vals = np.clip(x_vals * a + b, 1e-5, None)
y_fit = np.log(inner_vals) * c + d

# Plot 1: m, c, and MSE over iterations
plt.figure(figsize=(14, 5))

plt.subplot(1, 5, 1)
plt.plot(range(num_iterations), a_values)
plt.title('(a) over Iterations')
plt.xlabel('Iteration')
plt.ylabel('c value')

plt.subplot(1, 5, 2)
plt.plot(range(num_iterations), b_values)
plt.title('(b) over Iterations')
plt.xlabel('Iteration')
plt.ylabel('b value')

plt.subplot(1, 5, 3)
plt.plot(range(num_iterations), c_values)
plt.title('(c) over Iterations')
plt.xlabel('Iteration')
plt.ylabel('c value')

plt.subplot(1, 5, 4)
plt.plot(range(num_iterations), d_values)
plt.title('(d) over Iterations')
plt.xlabel('Iteration')
plt.ylabel('d value')

plt.subplot(1, 5, 5)
plt.plot(range(num_iterations), mse_values)
plt.title('MSE over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')

plt.tight_layout()
plt.show()

# Plot the fit
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.4, s=10, label="Diamonds Data")
plt.scatter(x_vals, y_fit, color='red', label="y = ln(x * a + b) * c + d")
plt.title("Logarithmic Fit to Diamonds Data")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

print
```
